```python
from pathlib import Path
import pandas as pd

RAW_DIR = Path (r"C:\Users\iberkayo\Desktop\turkish-appliance-rag-copilot\data\raw\Vestel")

rows = []

for category_dir in RAW_DIR.iterdir():
    if not category_dir.is_dir():
        continue

    category = category_dir.name

    for pdf_path in category_dir.glob("*.pdf"):
        rows.append({
            "brand": "Vestel",
            "category": category,
            "file_name": pdf_path.name,
            "local_path": str(pdf_path).replace("\\", "/"),
            "source_url": "",
            "model": "",
            "doc_type": "user_manual",
        })

df = pd.DataFrame(rows)

output_path = "C:/Users/iberkayo/Desktop/turkish-appliance-rag-copilot/data/raw/Vestel/corpus_manifest.xlsx"

df.to_excel(output_path, index=False)

print(df)
print(f"\nTotal PDFs: {len(df)}")
print(f"\nSaved to: {output_path}")
```
